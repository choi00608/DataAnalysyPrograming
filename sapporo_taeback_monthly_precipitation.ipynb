```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')
```

```python
jp_sapporo_13 = pd.read_csv("/content/jp_13.csv", encoding = 'cp949')
jp_sapporo_14 = pd.read_csv("/content/jp_14.csv", encoding = 'cp949')
jp_sapporo_15 = pd.read_csv("/content/jp_15.csv", encoding = 'cp949')
jp_sapporo_16 = pd.read_csv("/content/jp_16.csv", encoding = 'cp949')
jp_sapporo_17 = pd.read_csv("/content/jp_17.csv", encoding = 'cp949')
jp_sapporo_18 = pd.read_csv("/content/jp_18.csv", encoding = 'cp949')
jp_sapporo_19 = pd.read_csv("/content/jp_19.csv", encoding = 'cp949')
jp_sapporo_20 = pd.read_csv("/content/jp_20.csv", encoding = 'cp949')
jp_sapporo_21 = pd.read_csv("/content/jp_21.csv", encoding = 'cp949')
jp_sapporo_22 = pd.read_csv("/content/jp_22.csv", encoding = 'cp949')
jp_sapporo_23 = pd.read_csv("/content/jp_23.csv", encoding = 'cp949')

taeback = pd.read_csv("/content/taeback.csv", encoding = 'cp949')
taeback_feb = pd.read_csv("/content/taeback_feb.csv", encoding = 'cp949')

```

```python
jp_sapporo_13 = jp_sapporo_13.drop(jp_sapporo_13.index[jp_sapporo_13["강수량"] < 0])
jp_sapporo_14 = jp_sapporo_14.drop(jp_sapporo_14.index[jp_sapporo_14["강수량"] < 0])
jp_sapporo_15 = jp_sapporo_15.drop(jp_sapporo_15.index[jp_sapporo_15["강수량"] < 0])
jp_sapporo_16 = jp_sapporo_16.drop(jp_sapporo_16.index[jp_sapporo_16["강수량"] < 0])
jp_sapporo_17 = jp_sapporo_17.drop(jp_sapporo_17.index[jp_sapporo_17["강수량"] < 0])
jp_sapporo_18 = jp_sapporo_18.drop(jp_sapporo_18.index[jp_sapporo_18["강수량"] < 0])
jp_sapporo_19 = jp_sapporo_19.drop(jp_sapporo_19.index[jp_sapporo_19["강수량"] < 0])
jp_sapporo_20 = jp_sapporo_20.drop(jp_sapporo_20.index[jp_sapporo_20["강수량"] < 0])
jp_sapporo_21 = jp_sapporo_21.drop(jp_sapporo_21.index[jp_sapporo_21["강수량"] < 0])
jp_sapporo_22 = jp_sapporo_22.drop(jp_sapporo_22.index[jp_sapporo_22["강수량"] < 0])
jp_sapporo_23 = jp_sapporo_23.drop(jp_sapporo_23.index[jp_sapporo_23["강수량"] < 0])
```

```python


jp_sapporo =  pd.DataFrame({
    '일시' : [
    '2013-01', '2014-01', '2015-01', '2016-01', '2017-01', '2018-01', '2019-01', '2020-01', '2021-01', '2022-01', '2023-01'],

    '강수량' : [
        jp_sapporo_13.강수량.dropna().sum(),
        jp_sapporo_14.강수량.dropna().sum(),
        jp_sapporo_15.강수량.dropna().sum(),
        jp_sapporo_16.강수량.dropna().sum(),
        jp_sapporo_17.강수량.dropna().sum(),
        jp_sapporo_18.강수량.dropna().sum(),
        jp_sapporo_19.강수량.dropna().sum(),
        jp_sapporo_20.강수량.dropna().sum(),
        jp_sapporo_21.강수량.dropna().sum(),
        jp_sapporo_22.강수량.dropna().sum(),
        jp_sapporo_23.강수량.dropna().sum()]
               })

taeback = taeback.dropna()
jp_sapporo
```

```python
# @title 기본 제목 텍스트
fig, axs =  plt.subplots(figsize = (10, 10))

x1 = jp_sapporo.일시
y1 = jp_sapporo.강수량

x2 = taeback.일시
y2 = taeback.강수량

axs.plot(x1, y1, marker = 'o', color ="#FF92B1", label = 'SAPPORO')
axs.bar(x2, y2, label = 'TAEBACK', color = '#28B4B4')

axs.set_title('Amount of Precipitation, Every January')
axs.set_xlabel("Year")
axs.set_ylabel("precipitation(mm)")
plt.legend()

for i, v in enumerate(x1):
    plt.text(v, y1[i], y1[i],                 # 좌표 (x축 = v, y축 = y[0]..y[1], 표시 = y[0]..y[1])
             fontsize = 8,
             color='black',
             horizontalalignment='right',  # horizontalalignment (left, center, right)
             verticalalignment='bottom')    # verticalalignment (top, center, bottom)

for i, v in enumerate(x2):
    plt.text(v, y2[i], y2[i],                 # 좌표 (x축 = v, y축 = y[0]..y[1], 표시 = y[0]..y[1])
             fontsize = 8,
             color='blue',
             horizontalalignment='left',  # horizontalalignment (left, center, right)
             verticalalignment='bottom')    # verticalalignment (top, center, bottom)

plt.show()
```
